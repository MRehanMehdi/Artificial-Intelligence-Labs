```python
# append()

numbers = [1, 2, 3]
numbers.append(4)

print("After append:", numbers)
```

```python
# clear()

fruits = ["apple", "banana", "mango"]
fruits.clear()

print("After clear:", fruits)
```

```python
# copy()

original_list = [10, 20, 30]
copied_list = original_list.copy()

print("Original List:", original_list)
print("Copied List:", copied_list)
```

```python
# count()

values = [1, 2, 3, 2, 2, 4]
count_of_2 = values.count(2)

print("Count of 2:", count_of_2)
```

```python
# extend()

list1 = [1, 2, 3]
list2 = [4, 5, 6]

list1.extend(list2)

print("After extend:", list1)
```

```python
# index()

colors = ["red", "green", "blue", "green"]
index_of_green = colors.index("green")

print("Index of 'green':", index_of_green)
```

```python
# insert()

numbers = [1, 2, 4, 5]
numbers.insert(2, 3)

print("After insert:", numbers)
```

```python
# pop()

items = ["pen", "pencil", "eraser"]
removed_item = items.pop(1)

print("Removed Item:", removed_item)
print("After pop:", items)
```

```python
# remove()

animals = ["cat", "dog", "rabbit", "dog"]
animals.remove("dog")

print("After remove:", animals)
```

```python
# reverse()

numbers = [1, 2, 3, 4, 5]
numbers.reverse()

print("After reverse:", numbers)
```

```python
# sort()

numbers = [5, 2, 8, 1, 3]
numbers.sort()

print("After sort:", numbers)
```

```python
# GPA Calculator using Lists, Loop and If-Else

subjects = ["OOP", "DB", "Pak St.", "English"]
credits = [4, 3, 2, 2]

total_points = 0
total_credits = 0

for i in range(len(subjects)):
    marks = int(input("Enter marks for " + subjects[i] + ": "))

    if marks >= 85:
        gp = 4.0
    elif marks >= 75:
        gp = 3.5
    elif marks >= 65:
        gp = 3.0
    elif marks >= 55:
        gp = 2.5
    else:
        gp = 2.0

    total_points += gp * credits[i]
    total_credits += credits[i]

gpa = total_points / total_credits
print("Your GPA is:", round(gpa, 2))
```
